```python
#Import Dependencies

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import psycopg2
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect, func

#Create Engine/Connection to postgressql

engine = create_engine('postgres://postgres:postgres@localhost:5432/employees')
connection = engine.connect()
```

```python
# Check if engine/execute is running correctly by printing out data from departments table

data = engine.execute('SELECT * FROM departments LIMIT 5')
for record in data:
    print(record)
```

```python
# Set up inspector to read in table names and confirm they all have been connceted to via the engine

inspector = inspect(engine)

inspector.get_table_names()
```

```python
# View data in the salaries column to view formatting

engine.execute('SELECT * FROM salaries LIMIT 10').fetchall()
```

```python
# Set up refelction to pull table in to pandas and query

Base = automap_base()
Base.prepare(engine, reflect=True)
Salaries = Base.classes.salaries
```

```python
# Create a Session

session = Session(engine)
```

```python
# Query Salary column of Salaries table in desc order to confirm the highest salary in the table

results = session.query(Salaries.salary).order_by(Salaries.salary.desc()).all()

results = pd.DataFrame(results)

results.head()
```

```python
# Use Pandas to load a query statement directly into the DataFrame

stmt = session.query(Salaries).statement
df1 = pd.read_sql_query(stmt, session.bind)
df1
```

```python
# Create new datafram for salary

salary_df = df1[['emp_no', 'salary']]

salary_df
```

```python
# Create bins and labels to be used to be used for Salary Range

bins = [39999, 49999, 59999, 69999, 79999, 89999, 99999, 109999, 119999,
        129999, 150999]

labels = ["40K-49K", "50K-59K", "60K-69K", "70K-79K", "80K-89K", "90K-99K",
         "100K-109K", "110K-119K", "120K-129K", ">=130K"]
```

```python
# Slice bins together with salary_df

salary_df["salary_summary"] = pd.cut(salary_df["salary"], bins=bins, labels=labels)

salary_df
```

```python
# group data together by salary_summary and perform count funtion to get counts for each salary range

grouped_data = salary_df.groupby("salary_summary")

grouped_df = grouped_data["salary_summary"].count()

grouped_df = pd.DataFrame(grouped_df)

grouped_df = grouped_df.rename(columns={'salary_summary':'count'})

salaries_df = grouped_df.reset_index()

salaries_df
```

```python
# plot Salary Ranges for Employees table

plt.figure(figsize=(10,5))

y_axis = salaries_df['count'].tolist()

x_axis = np.arange(len(y_axis))

ticks = ["40K-\n49K", "50K-\n59K", "60K-\n69K", "70K-\n79K", "80K-\n89K", "90K-\n99K",
         "100K-\n109K", "110K-\n119K", "120K-\n129K", ">=130K"]

plt.bar(x_axis, y_axis, color='b', align='center')

tick_locations = [value for value in x_axis]

plt.xticks(tick_locations, ticks)

plt.title("Salary Ranges for Employees")

plt.xlabel("Salary")

plt.ylabel("Number of Employees")

plt.tight_layout()

#plt.savefig("../images/histogram.png")
```

```python
# Set up reflection for Titles table and read in sql query on table using pandas

Titles = Base.classes.titles

stmt = session.query(Titles).statement
df2 = pd.read_sql_query(stmt, session.bind)
df2
```

```python
# Merge salary table with titles table

merged_data = pd.merge(df2, df1, on='emp_no', how='inner', indicator=True)

merged_data.head(10)
```

```python
# Create new dataframe with the required fields for the next plot

merged_df = merged_data[['title', 'salary']]

merged_df
```

```python
# Group dataframe by title and run mean function on the dataframe to get the average salary per title

groupby_title = merged_df.groupby('title')

group_by = round(groupby_title["salary"].mean())

title_avg = pd.DataFrame(group_by)

title_avg_df = title_avg.reset_index()

title_avg_df
```

```python
# Plot bar chart of Average Salry by Title

plt.figure(figsize=(10,5))

y_axis = title_avg_df['salary'].tolist()

x_axis = np.arange(len(y_axis))

ticks = title_avg_df['title'].tolist()

plt.bar(x_axis, y_axis, color='b', align='center')

tick_locations = [value for value in x_axis]

plt.xticks(tick_locations, ticks)

plt.title("Average Salary by Title")

plt.xlabel("Title")

plt.ylabel("Average Salary")

plt.tight_layout()

#plt.savefig("../images/bar.png")
```
